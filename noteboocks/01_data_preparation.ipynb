```python
df_split = df.copy()

df_split[['freq','unit','airpol','src_crf','geo']] = df_split['freq,unit,airpol,src_crf,geo\\TIME_PERIOD'].str.split(',', expand=True)

df_split.head()
```

```python

df_split.columns

```

```python

paises = {
    "AT":"Austria","BE":"Belgium","BG":"Bulgaria","CY":"Cyprus","CZ":"Czech Republic",
    "DE":"Germany","DK":"Denmark","EE":"Estonia","EL":"Greece","ES":"Spain","FI":"Finland",
    "FR":"France","HR":"Croatia","HU":"Hungary","IE":"Ireland","IT":"Italy","LT":"Lithuania",
    "LU":"Luxembourg","LV":"Latvia","MT":"Malta","NL":"Netherlands","PL":"Poland","PT":"Portugal",
    "RO":"Romania","SE":"Sweden","SI":"Slovenia","SK":"Slovakia"
}
```

```python

df_split["country"] = df_split["geo"].map(paises)

df_split[["geo","country"]].head()
```

```python
df_split["geo"].unique()
```

```python

for pais in df_split["country"].unique():
    print(pais)
```

```python
df_split.columns.tolist()
```

```python
df_split.columns = df_split.columns.str.strip()
```

```python
df_split.columns.tolist()
```

```python
anos = df_split.loc[:, "1990":"2023"]

anos = anos.apply(pd.to_numeric, errors="coerce")

```

```python
df_split["total_emissoes"] = anos.sum(axis=1)

```

```python
ranking = (
    df_split
    .groupby("country")["total_emissoes"]
    .sum()
    .sort_values(ascending=False)
)

ranking.head()

```

```python

df_split.to_csv("nome_do_ficheiro.tsv", sep="\t", index=False)

```
