```python
def map_sector(code):
    if code.startswith("CRF1"):
        return "Energia"
    elif code.startswith("CRF2"):
        return "Processos Industriais"
    elif code.startswith("CRF3"):
        return "Agricultura"
    elif code.startswith("CRF4"):
        return "Uso do Solo e Florestas (LULUCF)"
    elif code.startswith("CRF5"):
        return "Resíduos"
    elif code.startswith("CRF6"):
        return "Outros"
    elif code.startswith("TOTX"):
        return "Memo"
    else:
        return "Outro"

df_split["sector_main"] = df_split["src_crf"].apply(map_sector)
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

# Top 10 países mais poluentes em 2023
df_2023 = df_long[df_long["year"] == 2023]

top10 = (df_2023.groupby("country")["emissions"]
         .sum()
         .sort_values(ascending=False)
         .head(10)
         .index)

plt.figure()

for country in top10:
    
    country_data = (df_long[df_long["country"] == country]
                    .groupby("year")["emissions"]
                    .sum()
                    .reset_index())
    
    X = country_data["year"].values.reshape(-1, 1)
    y = country_data["emissions"].values
    
    model = LinearRegression()
    model.fit(X, y)
    
    # Histórico até 2023
    plt.plot(country_data["year"], y)
    
    # Previsão apenas 2023–2030
    future_years = np.arange(2023, 2031).reshape(-1, 1)
    predictions = model.predict(future_years)
    
    plt.plot(future_years, predictions, linestyle="--", label=country)

plt.title("Top 10 Países — Histórico e Projeção (2023–2030)")
plt.xlabel("Ano")
plt.ylabel("Emissões")
plt.legend()
plt.show()

```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import numpy as np
import pandas as pd

df_2023 = df_long[df_long["year"] == 2023]

top5 = (df_2023.groupby("country")["emissions"]
        .sum()
        .sort_values(ascending=False)
        .head(5)
        .index)

results = []

for country in top5:
    
    country_data = (df_long[df_long["country"] == country]
                    .groupby("year")["emissions"]
                    .sum()
                    .reset_index())
    
    X = country_data["year"].values.reshape(-1, 1)
    y = country_data["emissions"].values
    
    model = LinearRegression()
    model.fit(X, y)
    
    y_pred = model.predict(X)
    
    slope = model.coef_[0]
    r2 = r2_score(y, y_pred)
    
    results.append({
        "country": country,
        "slope (tendência anual)": slope,
        "R² (qualidade do ajuste)": r2
    })

regression_summary = pd.DataFrame(results).sort_values("slope (tendência anual)")

regression_summary
```
